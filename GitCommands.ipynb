```
import ipdb
class Solution:
    def lengthOfLongestSubstring(self, s):
        # ipdb.set_trace()
        dicts = {}
        maxlength = start = 0
        for i,value in enumerate(s):
            if value in dicts:
                ipdb.set_trace()
                sums = dicts[value] + 1
                if sums > start:
                    start = sums
            num = i - start + 1
            if num > maxlength:
                maxlength = num
            dicts[value] = i
        return maxlength
input_string = "pwwkew"
a  = Solution()
answer = a.lengthOfLongestSubstring(input_string)
print(answer)
```
